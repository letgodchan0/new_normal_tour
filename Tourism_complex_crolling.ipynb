```python
import selenium
from selenium import webdriver
# from selenium.webdriver import ActionChains
from bs4 import BeautifulSoup
import time
import re
```

```python
URL = 'https://api.visitkorea.or.kr/search/commonList.do'

관광지명 = []
지역 = []
cnt = 0
# 홈페이지 열기
driver = webdriver.Chrome(executable_path='./chromedriver')
driver.get(url=URL)
driver.implicitly_wait(time_to_wait=3)

# 쇼핑, 음식점 카테고리 제거하기
driver.find_element_by_xpath('//*[@id="typeid_38"]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="typeid_39"]').click()
driver.implicitly_wait(time_to_wait=3)
time.sleep(2)

# 자연 먼저 확인하기
driver.find_element_by_xpath('//*[@id="selectDiv"]/table/tbody/tr[2]/td/div[1]/span[1]/select[1]/option[3]').click()
# //*[@id="selectDiv"]/table/tbody/tr[2]/td/div[1]/span[1]/select[1]
driver.implicitly_wait(time_to_wait=3)
time.sleep(2)


# 메뉴 50개씩 보기
driver.find_element_by_xpath('//*[@id="numOfPage"]/option[5]').click()
# //*[@id="numOfPage"]
driver.implicitly_wait(time_to_wait=1)
driver.find_element_by_xpath('//*[@id="numOfPageView"]').click()
# //*[@id="numOfPageView"]
driver.implicitly_wait(time_to_wait=5)

while True:
    # 페이지 넘어가기 알고리즘
    try:
        if cnt >= 1:
            driver.find_element_by_xpath('//*[@id="content"]/div[6]/span/a[{}]'.format(cnt)).click()
            driver.implicitly_wait(time_to_wait=3)
            time.sleep(2)
        cnt += 1
    except:
        break
        
    if cnt == 10:
        driver.find_element_by_xpath('//*[@id="content"]/div[6]/a[3]').click()
        driver.implicitly_wait(time_to_wait=3)
        time.sleep(2)
        cnt = 0

    # 해당 페이지 소스 받아오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    attraction = soup.select('#listForm > ul > li > a > dl > dt')
#   //*[@id="listForm"]/ul/li[1]/a/dl/dt
    location = soup.select('#listForm > ul > li > a > dl > dd > strong')
#   //*[@id="listForm"]/ul/li[1]/a/dl/dd/strong
    for i in range(len(attraction)):
        관광지명.append(attraction[i].text)
        지역.append(location[i].text)
관광지명
```
